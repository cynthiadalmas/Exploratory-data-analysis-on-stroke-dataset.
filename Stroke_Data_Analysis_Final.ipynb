```python
#checking for outlier
# a boxplot to visualize distribution of any outliers in age
#plt.subplot(1,3,1)
plt.figure(figsize=(3,2))
plt.title('Boxplot to detect outliers for age', fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
sns.boxplot(x=df1['age'], palette = 'Set2')


# a boxplot to visualize distribution of any outliers in Avg_glucose_level
#plt.subplot(1,3,2)
plt.figure(figsize=(3,2))
plt.title('Boxplot to detect outliers for avg_glucose_level', fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
sns.boxplot(x=df1['avg_glucose_level'], palette = 'Set2')


# a boxplot to visualize distribution of any outliers in BMI
#plt.subplot(1,3,3)
plt.figure(figsize=(3,2))
plt.title('Boxplot to detect outliers for bmi', fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
sns.boxplot(x=df1['bmi'], palette = 'Set2')
plt.show()

```

```python
#handling the outlier

Q1=df1['bmi'].quantile(0.25)
Q3=df1['bmi'].quantile(0.75)


# Computing the interquartile range
iqr =Q3-Q1

#Define the upper and lower limits
upper_limit =Q3+1.5 *iqr
lower_limit =Q1-1.5 *iqr

#identify the outlier
outlier_bmi=df1[(df1['bmi'] > upper_limit) | (df1['bmi'] < lower_limit)] 
```

```python
#handling the outlier

Q1=df1['avg_glucose_level'].quantile(0.25)

Q3=df1['avg_glucose_level'].quantile(0.75)


# Computing the interquartile range
iqr =Q3-Q1

#Define the upper and lower limits
upper_limit =Q3+1.5 *iqr
lower_limit =Q1-1.5 *iqr

#identify the outlier
outlier=df1[(df1['avg_glucose_level'] > upper_limit) | (df1['avg_glucose_level'] < lower_limit)] 
```
